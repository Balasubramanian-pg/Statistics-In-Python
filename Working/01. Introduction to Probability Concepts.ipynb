## Core Concepts

### 1. Probability Axioms

Probability is governed by three foundational axioms introduced by Andrey Kolmogorov.

#### Axiom 1: Non-Negativity
The probability of any event is always greater than or equal to zero.

$$P(A) \geq 0$$

#### Axiom 2: Normalization
The probability of the sample space equals 1.

$$P(S) = 1$$

#### Axiom 3: Additivity
If two events are mutually exclusive:

$$P(A \cup B) = P(A) + P(B)$$


```python
# Example: Rolling a Die

sample_space = [1, 2, 3, 4, 5, 6]
even_numbers = [2, 4, 6]

prob_even = len(even_numbers) / len(sample_space)

print(f'Probability of rolling an even number: {prob_even}')
```

### 2. Independent vs Dependent Events

Two events are independent if the occurrence of one does not affect the probability of the other.

Examples:
- Tossing a coin and rolling a die
- Selecting cards with replacement

Dependent events influence each other.

Examples:
- Drawing cards without replacement
- Weather conditions affecting traffic


```python
# Independent Event Example

p_heads = 1/2
p_six = 1/6

p_both = p_heads * p_six

print(f'Probability of Heads and a 6: {p_both:.4f}')
```

### 3. Marginal Probability

Marginal probability measures the probability of a single event occurring regardless of other variables.


```python
# Marginal Probability Example

data = pd.DataFrame({
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'Purchased': ['Yes', 'No', 'Yes', 'Yes', 'No']
})

p_purchase = (data['Purchased'] == 'Yes').mean()

print(f'Marginal Probability of Purchase: {p_purchase:.2f}')
```

### 4. Complementary Events

The complement of an event represents outcomes where the event does not occur.

$$P(A^c) = 1 - P(A)$$


```python
# Complementary Event Example

p_rain = 0.3
p_no_rain = 1 - p_rain

print(f'Probability of no rain: {p_no_rain}')
```

## Simulation Example: Law of Large Numbers

As the number of experiments increases, empirical probability approaches theoretical probability.


```python
# Simulating Coin Tosses

coin_tosses = np.random.choice(['Heads', 'Tails'], size=1000)

heads_count = np.cumsum(coin_tosses == 'Heads')
running_probability = heads_count / np.arange(1, 1001)

plt.figure(figsize=(12, 6))
plt.plot(running_probability)
plt.axhline(y=0.5, linestyle='--')

plt.title('Law of Large Numbers')
plt.xlabel('Number of Tosses')
plt.ylabel('Probability of Heads')

plt.show()
```

## Dice Roll Simulation

Empirical probability can be estimated using repeated experiments.


```python
# Dice Roll Simulation

rolls = np.random.randint(1, 7, size=10000)

unique, counts = np.unique(rolls, return_counts=True)
probabilities = counts / len(rolls)

plt.figure(figsize=(10, 5))
plt.bar(unique, probabilities)
plt.axhline(y=1/6, linestyle='--')

plt.title('Empirical Probability of Dice Rolls')
plt.xlabel('Dice Face')
plt.ylabel('Probability')

plt.show()
```
